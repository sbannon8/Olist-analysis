### Re-Categorize Product Categories

```python
# define new catories & the existing categories they will encompass
custom_category_mapping = {
    'Fashion': ['fashion_sport', 'fashion_shoes', 'fashion_male_clothing', 'fashion_childrens_clothes', 'fashion_underwear_beach', 'fashio_female_clothing', 'fashion_bags_accessories', 'luggage_accessories', 'sports_leisure'],
    'Beauty': ['perfumery', 'health_beauty'],
    'Art': ['arts_and_craftmanship',  'art'],
    'Music & Film': ['music', 'cds_dvds_musicals', 'cine_photo', 'dvds_blu_ray', 'dvds_blu_ray', 'musical_instruments'],
    'Technology': ['tablets_printing_image',  'electronics', 'consoles_games', 'fixed_telephony', 'audio', 'computers_accessories', 'telephony', 'computers'],
    'Books & Stationary': ['stationery', 'books_general_interest', 'books_imported', 'books_technical'],
    'Home & Garden': ['bed_bath_table', 'housewares', 'furniture_living_room', 'furniture_decor', 'small_appliances', 'home_confort', 'air_conditioning', 'home_appliances_2', 'kitchen_dining_laundry_garden_furniture', 'small_appliances_home_oven_and_coffee', 'furniture_bedroom', 'furniture_mattress_and_upholstery',  'home_comfort_20', 'home_appliances', 'garden_tools', 'flowers', 'pet_shop', 'office_furniture'],
    'Food & Drink': ['food', 'food_drink', 'la_cuisine', 'drinks'],
    'Baby & Kids': ['baby', 'diapers_and_hygiene', 'toys'],
    'Gift': ['cool_stuff', 'christmas_supplies', 'party_supplies', 'watches_gifts'],
    'Auto': ['auto'],
    'Security': ['signaling_and_security', 'security_and_services'],
    'Construction': ['home_construction', 'costruction_tools_garden', 'construction_tools_construction', 'construction_tools_safety', 'construction_tools_lights', 'costruction_tools_tools'],
    'Commerce': ['market_place', 'industry_commerce_and_business', 'agro_industry_and_commerce'],
}

# map existing categories to custom categories
def map_to_custom_category(category):
    for custom_category, categories in custom_category_mapping.items():
        if category in categories:
            return custom_category
    return 'Other'  # if doesn't match any, assign it to 'Other' (for the nan values)

# apply mapping to df
df['custom_category'] = df['product_category_name_english'].apply(map_to_custom_category)
```
